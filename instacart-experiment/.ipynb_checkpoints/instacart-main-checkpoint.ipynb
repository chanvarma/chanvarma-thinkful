Data dictionary from https://gist.github.com/jeremystan/c3b39d947d9b88b3ccff3147dbcf6c6b



`orders` (3.4m rows, 206k users):
* `order_id`: order identifier
* `user_id`: customer identifier
* `eval_set`: which evaluation set this order belongs in (see `SET` described below)
* `order_number`: the order sequence number for this user (1 = first, n = nth)
* `order_dow`: the day of the week the order was placed on
* `order_hour_of_day`: the hour of the day the order was placed on
* `days_since_prior`: days since the last order, capped at 30 (with NAs for `order_number` = 1)

`products` (50k rows):
* `product_id`: product identifier
* `product_name`: name of the product
* `aisle_id`: foreign key
* `department_id`: foreign key

`aisles` (134 rows):
* `aisle_id`: aisle identifier
* `aisle`: the name of the aisle

`deptartments` (21 rows):
* `department_id`: department identifier
* `department`: the name of the department

`order_products__SET` (30m+ rows):
* `order_id`: foreign key
* `product_id`: foreign key
* `add_to_cart_order`: order in which each product was added to cart
* `reordered`: 1 if this product has been ordered by this user in the past, 0 otherwise

where `SET` is one of the four following evaluation sets (`eval_set` in `orders`):
* `"prior"`: orders prior to that users most recent order (~3.2m orders)
* `"train"`: training data supplied to participants (~131k orders)
* `"test"`: test data reserved for machine learning competitions (~75k orders)

## Business case

Assume that you work on Instacart's Data Science team, and one of your biggest sellers, a major soft drink company, approaches you with a problem. Their **line of soda brands is losing market share in the American market to other beverage categories**, and the company believe this to be a failure in understanding changing consumption patterns. They want some insight into the consumption patterns of beverages to better understand this issue. 

Moreover, the company is planning to **develop and introduce a new non-soft-drink product to compete in the changing market, and plans to beta-test it exclusively on Instacart** by displaying ads at an opportune time. Set up a framework for carrying out this experiment, with clearly defined problem statements, control/treatment segmentation, success variables, and other concerns.

## Approach

### Building customer insight
1. Isolate order data to only those containing orders with products from the `beverages` aisle. 
2. Within this data, understand how orders containing `soft drinks` are different from orders not containing `soft drinks`.
3. Build a theory for consumption patterns. 

### Building experimentation framework
1. Based on part 2.1, build hypothesis for introducing the new product.

To simplify our analysis, we're going to create a `master` dataset that includes complete information for every product in every order. 

```python
master = orders_sample.merge(order_products__prior, on = 'order_id', how = 'inner')
master.head()
```

We also create two helper functions to signify which orders (a) contain beverages, and (b) specifically contain sodas.

```python
soft_drink_product_ids = list(products['product_id'].loc[products['aisle_id'] == 77]) ## 77: soft drinks
bevs_product_ids = list(products['product_id'].loc[products['department_id'] == 7]) #dept_id 7: beverages

def contains_soft_drink(product_id):
    return product_id in soft_drink_product_ids

def contains_bevs(product_id):
    return product_id in bevs_product_ids
```

```python
master['contains_soft_drinks'] = master['product_id'].apply(contains_soft_drink)
master['contains_bevs'] = master['product_id'].apply(contains_bevs)

bevs_only = master[master['contains_bevs'] == True]
bevs_only = bevs_only.merge(products, on = 'product_id')
bevs_only = bevs_only.merge(aisles, on = 'aisle_id', how = 'left')
bevs_only = bevs_only.loc[bevs_only['aisle'] != 'cocoa drink mixes']
bevs_only.sort_values(by = 'order_id', inplace = True)
bevs_only['counter'] = 1
bevs_only.head()
```

## Visualising the beverage "aisle"

### Visualising the distribution of the different types of beverages

```python
def aisle_rename(aisle):
    if aisle in ['tea', 'coffee']:
        return 'tea/coffee-based drinks'
    if aisle == 'water seltzer sparkling water':
        return 'water/seltzer'
    return aisle

bevs_only['aisle'] = bevs_only['aisle'].apply(aisle_rename)
vc = bevs_only['aisle'].value_counts()

sns.barplot(y = vc.index, x = vc.values)
plt.show()
```

We plot the bar charts for each of these aisles, to better understand how the popularity of individual products is distributed. Since there are over 12000+ unique products, we'll just focus on the top 50 for now. To standardize our analysis, we'll plot the ratio of products/total products for each `product_id`.

```python
names_of_aisles = bevs_only['aisle'].unique()
```

```python
for index, aisle in enumerate(names_of_aisles):
    f = plt.figure(figsize = (10, 1.5))
    
    vc = bevs_only[bevs_only['aisle'] == aisle]['product_name'].value_counts()
    total = sum(vc.values)
    
    title = "Distribution of 50 most popular '{}' products".format(aisle)
    sns.barplot(x = vc[:50].index, y = vc[:50].values/total).set_title(title)
    plt.axhline(y = np.mean(vc.values)/total, label = 'mean of distribution', c = 'red', lw = 1)
    plt.axhline(y = np.median(vc.values)/total, label = 'median of distribution', c = 'green', lw = 1)
    plt.axvline(x = 10, label = '10th most popular product', c = 'blue', lw = 1)
    plt.axvline(x = 20, label = '20th most popular product', c = 'orange', lw = 1)
    
    plt.legend(fontsize = 'x-small')
    plt.ylim(0, .1)
    
    cur_axes = plt.gca()
    cur_axes.axes.get_xaxis().set_ticklabels([])
    
plt.show()
```

This is a lot more revealing! It's quite evident that some aisles are far more left-skewed than the others. A greater left-skew simply means that the most popular products enjoy some of the highest market shares, whereas the flatter distributions (such as the one seen in tea/coffee products) indicates weaker penetration by a single product. 

### Estimating market share and competitiveness of various `aisles`

I wanted to have a more formal understanding of how market share varied across the different aisles for the top 'n' products, so I created a helper function.

```python
top_products = dict()
def easy_percentage(vc_series, n):
    return round(sum(vc_series[:n].values)/sum(vc_series.values), 2)


for aisle in names_of_aisles:
    vc = bevs_only[bevs_only['aisle'] == aisle]['product_name'].value_counts()

    top_products[aisle] = {'number_of_products': len(vc),
                           'number_of_products_ordered': sum(vc.values),
                           'top_5_products_comprise': easy_percentage(vc, 5),
                           'top_10_products_comprise': easy_percentage(vc, 10),
                           'top_25_products_comprise': easy_percentage(vc, 25),
                           'top_50_products_comprise': easy_percentage(vc, 50),
                           'top_100_products_comprise': easy_percentage(vc, 100)}


pd.DataFrame.from_dict(top_products)
```

This confirms my hypothesis. The soda and seltzer aisles are dominated by a few brands, and over 50% of the market in almost every aisle (except tea/coffee) is owned by 25 brands. 

If the client is losing market share in a highly competitive market like this, they are more likely to lose large profits at a very quick rate. To better understand dominant consumption patterns, we'll create a subset of the top 25 brands in each aisle. 

```python
top25_productnames = []

for aisle in names_of_aisles:
    vc = bevs_only[bevs_only['aisle'] == aisle]['product_name'].value_counts()[
        :20]

    top25_productnames.extend(vc.index)

top25 = bevs_only.loc[bevs_only['product_name'].isin(top25_productnames)]

top25['product_name'].value_counts()[:15]
```

### Plotting order distributions across day-of-week and time-of-day for the top 3 aisles

We will also narrow our analysis to what seems like the three most competitive markets: water/seltzer, energy sports drinks, and refrigerated. All of these beverage types suggest carbonation, and are likely to be fighting for the same market share. It seems unlikely that a tea-based drink is taking over soda in large proportions. In the event that a chain of especially popular tea drinks (for example) were taking over the market, we would see it in the market share comparisions.

```python
main_aisles = ['soft drinks', 'water/seltzer', 'refrigerated']
```

#### Day of week distribution 

To compare across aisles, we divide the y-axis values by the sum of total values.

```python
x = top25.groupby(by=['aisle', 'order_dow'])['counter'].count()
x = x.unstack()
pd.DataFrame(x)

plt.figure(figsize=(8, 5))

for a in main_aisles:
    xs = x.xs(a)
    sns.lineplot(x=xs.index, y=xs.values)
    sns.scatterplot(x=xs.index, y=xs.values, label=a)

title = "Day-of-week order distribution of top 25 products in each category (unnormalised)"
plt.title(title, fontsize='x-large')
plt.show()

plt.figure(figsize=(8, 5))

for a in main_aisles:
    xs = x.xs(a)
    total = sum(xs.values)
    sns.lineplot(x=xs.index, y=xs.values/total)
    sns.scatterplot(x=xs.index, y=xs.values/total, label=a)

title = "Day-of-week order distribution of top 25 products in each category (normalised)"
plt.title(title, fontsize='x-large')
plt.show()
```

Aha! We discover something here. While `soft drinks` are the beverage of choice in the second half of the week, `water/seltzer` and `refrigerated beverages` reign supreme in the first half. When contextualized to the unweighted numbers in the chart above, this jump seems to be massive enough to shift sales.

#### Time of day distribution 

We do the same for the hour of the day, to see if there's something relevant.

```python
x = top25.groupby(by=['aisle', 'order_hour_of_day'])['counter'].count()
x = x.unstack()
pd.DataFrame(x)

plt.figure(figsize=(10, 5))
plt.xlim(0, 24)
plt.xlabel(range(23, 2))

for a in main_aisles:
    xs = x.xs(a)
    sns.lineplot(x=xs.index, y=xs.values)
    sns.scatterplot(x=xs.index, y=xs.values, label=a)
plt.axvline(x = 12, label = '12pm/midday', c= 'grey')

title = "Hour-of-day order distribution of top 25 products in each category (unnormalised)"
plt.title(title, fontsize='x-large')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))

for a in main_aisles:
    xs = x.xs(a)
    total = sum(xs.values)
    sns.lineplot(x=xs.index, y=xs.values/total)
    sns.scatterplot(x=xs.index, y=xs.values/total, label=a)

plt.axvline(x = 12, label = '12pm/midday', c= 'grey')
title = "Hour-of-day order distribution of top 25 products in each category (normalised)"
plt.title(title, fontsize='x-large')
plt.legend()
plt.show()
```

Unfortunately, the weighted graphs here are not very revealing. There seems to be a general dip in orders around noon, and it seems as though the orders for `water/seltzer` drop sharply in the evening. If anything, this suggests that demand for water/seltzer based products drops after the afternoon. 

Lastly, we summarise our top 25 tables to validate our visual findings.

```python
top25summary = top25.groupby(by='aisle').mean()[['order_hour_of_day', 'order_dow',
                                  'days_since_prior_order', 'add_to_cart_order', 'reordered']]


top25summary['total_products_sold'] = top25.groupby(by='aisle').sum()['counter']
top25summary['ratio_of_products_of_top25'] = top25summary['total_products_sold']/sum(top25summary['total_products_sold'])


top25summary
```

## Conclusions: Summary of Consumer Patterns


1. The market for beverages is dominated by 3 `aisles`: water/selzter, soft drinks, refrigerated. They cover over 77% of the top 150 products sold. 
2. Within these segments, market share is dominated by select few brands, with 25 brands covering over 50% of the market. 
3. Soft drinks are more likely to be ordered in the second half of the week, whereas water/seltzer dominates the opening half. I think this trend is likely to be caused as people tend to order soft drinks as beverage options for parties/get-togethers, and these events happen disproportionately in the second half of the week. 
4. In similar vein, seltzer/water is more likely to be order in between 10am and 3pm (and suffers a steep decline after that), whereas soft drinks and refrigerated drinks share a more gentle decline into the evening. 


## Recommendation to soft drink company
Based on everything, we conclude that the sales of soft drinks are being taken over by water/seltzer sales. This is indicative of a larger global trend towards healthier lives, and opting for drinks with a reduced/no sugar content is a big part of it. We recommend  that the soft drink company **launch a new seltzer product line**.

The dominant market share in this space presents a unique opportunity for the established soft drink company. By having existing and well-developed manufacturing and distribution channels, the soft drink company is well positioned to pivot into a market-leading position... if it has a product that fares well in a pilot test.


## Experimental design for new product pilot test
Based on the data around favorable times to order for water/selzters, **the 10am - 1pm window on Sundays and Mondays** seems especially favorable. We recommend beta-testing a new product during this window. 

To better understand the water/seltzer space, we look at some of the key values associated with the top 25 products in the space. 

```python
vars_of_interest = ['order_number', 'order_dow', 'order_hour_of_day', 'days_since_prior_order', 'add_to_cart_order']

top25.loc[top25['aisle'] == 'water/seltzer'].groupby('product_name')[vars_of_interest].mean()
```

Each of these values offers different variables for multi-variate testing. 

## Research design

### Problem
How can we best test the demand of a new beverage product? Soda sales are shrinking across the market and it's imperative to explore other revenue streams. The soda company should leverage existing strengths in distribution channels, manufacturing and bottling plants, and name-brand recognition.

### Potential solution
Adding a fruit-flavored seltzer will allow the company to test the water and hopefully win back some market share.

### Business objective
Breach the list of top 25 seltzer products sold for the particular market when the treatment group is compared with the control group.

### Pre-experiment work
Ask Soda company's R&D to create limited quantity of 3 new fruit flavored seltzers, and then identify key geographical markets to run the pilot test in. Care must be taken to not choose uniquely large urban markets (NYC, LA, SF etc.) or largely rural areas they might distort/inflate the actual demand. We recommend a "second-tier" city, such as one of the [following](https://meetingsource.com/second-tier-convention-cities/):
* Austin, TX
* Pittsburgh, PA
* Denver, CO

Agreements with supermarkets must be also be done in advance to store the inventory, and understand the pricing of the new product. I recommend pricing at a slight discount when compared to key competitors. 

### Sample
Based on the data we have, our sample will be 36% of the following population:
* anyone ordering on Instacart in the geography of choice 
* ordering in the first half of the week (Sunday -- Wednesday) 
* Has at least 3 **non-water/selzter** items in cart
* Has a history of buying water/selzter products (essentially, the `days_since_prior_order` > 0)

Of this sample, 9% of the population will be shown no advertisement (control group), and the remaining 27% will see 1 of the 3 variations of the new fruit-flavored drink. 

### Location of ad
The ad will be shown as a modal on the check-out page, i.e, when the user is reviewing their cart and is likely to place the order.

### Outcomes of interest
The following variables will be measured and compared across the control group and 3 treatment groups:
* CTR of advertisement 
* Add-to-bag rate of advertised product 
* Quantity of advertised product purchased
* Number of total new products added after advertisement shown

There are other variables associated with an order that are already being measured (eg. time, location, day of week, etc.) that will guide post-experiment analysis. As mentioned in the business objective, we will also be measuring the list of top 25 water/seltzer products sold for the sample.


### Goal
1. Confirm hypothesis and have at least 1 new flavor breach into the top 25 water/seltzer products for the sample.
2. Identify which flavor has the highest demand of the 3. 
2. Better understand the demographics and consumption habits of the customers ordering the best-selling flavor to guide future marketing roll out -- especially repeat customers. 


### Duration and future plans
The initial phase of this test will last for 1 month. If the goals are met, secondary phases of this experiment will include advertising to customers with soda products in their cart at non-optimal times, etc.

### Key concerns
* Discovery of the product by the media/social media might artificially inflate demand.
* Instacart drivers and shoppers will have to familiarize themselves with the launch of a new product.
