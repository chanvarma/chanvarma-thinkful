```python
k_means = KMeans(n_clusters=2, random_state=13)
k_means.fit(X)
y_pred = k_means.predict(X)

sum(y_pred == Y)/len(y_pred == Y)
```

Looks like the initial run was reasonably successful. To visualize the clustering, we'll run PCA and map the first two principal components.

```python
pca = PCA(n_components=2).fit_transform(X)

plt.figure(figsize=(10,5))
colours = 'rb'
for i in range(pca.shape[0]):
    plt.text(pca[i, 0], pca[i, 1], str(y_pred[i]),
             color=colours[y[i]],
             fontdict={'weight': 'bold', 'size': 75}
        )

plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```
