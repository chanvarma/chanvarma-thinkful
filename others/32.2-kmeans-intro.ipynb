```python
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_palette('deep')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import MiniBatchKMeans
```

### Preparing the data

```python
# Define the features and the outcome
X = raw.iloc[:, :13]
Y = raw.iloc[:, 13]

# Replace missing values (marked by ?) with a 0
X = X.replace(to_replace='?', value=0)

# Binarize y so that 1 means heart disease diagnosis and 0 means no diagnosis
Y = np.where(Y > 0, 0, 1)

X.head()
```

```python
Y[:5]
```

### Initial K-means

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)

pd.DataFrame(X).head()
```

```python
k_means = KMeans(n_clusters=2, random_state=13)
k_means.fit(X)
y_pred = k_means.predict(X)

sum(y_pred == Y)/len(y_pred == Y)
```

Looks like the initial run was reasonably successful. To visualize the clustering, we'll run PCA and map the first two principal components.

```python
pca = PCA(n_components=2).fit_transform(X)

plt.figure(figsize=(10,5))
colours = 'rb'
for i in range(pca.shape[0]):
    plt.text(pca[i, 0], pca[i, 1], str(y_pred[i]),
             color=colours[Y[i]],
             fontdict={'weight': 'bold', 'size': 75}
        )

plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

This is what an ideal solution looks like. We'll now run KMeans without a specified `n_cluster` parameter.

```python
kmeans2 = KMeans(random_state=13)
kmeans2.fit(X).n_clusters
```

Wow, looks like there's likely a greater degree of specificity as to whether or not the patients have heart disease. Let's also try and visualise this. 

```python
y_pred2 = kmeans2.predict(X)
```

```python
y_pred2
```

```python
pca = PCA(n_components=2).fit_transform(X)

plt.figure(figsize=(10,5))
colours = ['red', 'blue', 'green', 'orange', 'pink', 'black', 'purple', 'brown']
for i in range(pca.shape[0]):
    plt.text(pca[i, 0], pca[i, 1], str(y_pred2[i]),
             fontdict={'weight': 'bold', 'size': 75},
             color = colours[y_pred2[i]]
        )

plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

Of course, the real test is matching whether the 8 clusters represent subclusters in the initial clustering. We plot that...

```python
pca = PCA(n_components=2).fit_transform(X)

plt.figure(figsize=(10,5))
colours = 'rb'
for i in range(pca.shape[0]):
    plt.text(pca[i, 0], pca[i, 1], str(y_pred2[i]),
             color=colours[Y[i]],
             fontdict={'weight': 'bold', 'size': 75}
        )

plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

```python
for i in np.unique(y_pred2):
    print("For cluster {}, there is a variance of {}.".format(i, np.round(np.var(y_pred[y_pred2 == i]), 3)))
```

### Question 1
Split the data randomly into two and apply k-means using two, three and four as the number of clusters as we did in our discussion of the consistency in this checkpoint. Assess the consistency of the solutions using visualization. Which one does seem to be the best one?

```python
from sklearn.model_selection import train_test_split

X_half1, X_half2, X_pcahalf1, X_pcahalf2 = train_test_split(
    X, pca, test_size=0.5, random_state=13, stratify = Y)

X_half1, X_half2, X_pcahalf1, X_pcahalf2 = train_test_split(
    X, pca, test_size=0.5, random_state=13, stratify = Y)

X_half1.shape, X_half2.shape, X_pcahalf1.shape, X_pcahalf2.shape
```

```python
models = {}

for i in range(2, 5):
    kmeans1 = KMeans(n_clusters=i, random_state=13)
    kmeans1.fit(X_half1)
    y_pred1 = kmeans1.predict(X_half1)

    kmeans2 = KMeans(n_clusters=i, random_state=13)
    kmeans2.fit(X_half2)
    y_pred2 = kmeans2.predict(X_half2)
    
    models[i] = {'mod1': kmeans1, 'mod2': kmeans2}

    df1 = pd.DataFrame(X_pcahalf1, columns=['pca1', 'pca2'])
    df1['group'] = 1
    df1['y_pred'] = y_pred1

    df2 = pd.DataFrame(X_pcahalf2, columns=['pca1', 'pca2'])
    df2['group'] = 2
    df2['y_pred'] = y_pred2

    df_cluster = pd.concat([df1, df2], axis=0)
    df_cluster.reset_index(inplace=True, drop=True)

    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    #plt.figure(figsize=(12, 5))

    for g in [1, 2]:
        subset = df_cluster.loc[df_cluster['group'] == g]
        f = sns.scatterplot(x='pca1', y='pca2', data=subset,
                        hue='y_pred', ax=ax[g - 1], palette='deep')
        
        f.axhline(0, c = 'grey', lw = 0.5)
        f.axvline(0, c = 'grey', lw = 0.5)

    plt.suptitle("n_clusters = {}".format(i))
    plt.show()
```

Ignoring the order/color of the clustering, it appears that the clusters are clustered somewhat similarly across the two groups. Of the three, it definitely appears that `n_clusters` = 2 is the most consistent, but this could be because of visual bias.

### Question 2

Apply k-means on the whole dataset by setting k is equal two, three and four and get ARI score for each of them. Which model is the best?

```python
from sklearn.metrics import adjusted_rand_score, silhouette_score

for key in models.keys():
    for m in ['mod1', 'mod2']:
        
        model = models[key][m]
        full_pred = model.fit_predict(X)
        full_pred = np.array([0 if x == 1 else 1 if x==0 else 2 for x in full_pred])
        
        print("ARI score:", key, m, adjusted_rand_score(Y, full_pred))
        
for key in models.keys():
    for m in ['mod1', 'mod2']:
        
        model = models[key][m]
        full_pred = model.fit_predict(X)
        full_pred = np.array([0 if x == 1 else 1 if x==0 else 2 for x in full_pred])
        
        print("Silhouette coeff:", key, m, silhouette_score(X, full_pred, metric = 'euclidean'))
```

As expected, the 2 cluster solution has the highest ARI score and silhouette coefficient values.

Produce dendrograms for the heart disease dataset using three different linkage methods: `complete`, `average` and `ward`. Which linkage method do you think produces more reasonable result?

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.decomposition import PCA
```

```python
def show_dend(X_std, linkage_method):
    plt.figure(figsize=(20,10))
    dendrogram(linkage(X_std, method=linkage_method), no_labels=True)
    plt.show()
```

```python
for linkage_method in ['complete', 'ward', 'average']:
    print("Method:", linkage_method)
    show_dend(X, linkage_method)
```

Of the 3, it appears that the 'ward' method produces the most reasonable distribution, because it is somewaht evenly spaced out, matches most closely with the true values from n_clusters (=2).
