## Dealing with missing values and outliers

### Missing values

```python
data.isnull().sum()/data.isnull().count()
```

Clearly, filling in the missing values with the population median/means will distort the data, so we will interpolate using a specific `STATE`'s data. The remaining NaN values will be filled in with the mean of the year before and after. 

```python
data.loc[data['STATE'] == 'CALIFORNIA'].head(10)
```

```python
data2 = data.copy()
print(id(data), id(data2))
```

```python
years = list(data2['YEAR'].unique())
years.sort(reverse = True)
```

```python
for state in data2['STATE'].unique():
    data2.loc[data2['STATE'] == state] = data2.loc[data2['STATE'] == state].interpolate()
    
data2['YEAR'].loc[data2.isna().any(axis = 1)].value_counts()
```

Most of the missing values seem to be in the first few years of the dataset. We will now try and fill the missing values using the averages of the years before and after a given year.

```python
for state in data2['STATE'].unique():
    for year in years:
        cols_with_nans = data2.loc[(data2['YEAR'] == year) & (
            data2['STATE'] == state)].isna().any()
        cols_with_nans = list(cols_with_nans.loc[cols_with_nans == True].index)

        try:
            if len(cols_with_nans) > 0:
                for col in cols_with_nans:
                    if year > 1992 & year < 2017:  # first year on record
                        before = data2[col].loc[(
                            data2['YEAR'] == year - 1) & (data2['STATE'] == state)].values[0]
                        after = data2[col].loc[(
                            data2['YEAR'] == year + 1) & (data2['STATE'] == state)].values[0]

                        before_after_mean = np.nanmean([before, after])

                        data2[col].loc[(data2['YEAR'] == year) & (data2['STATE']
                                                                      == state)] = before_after_mean
                    if year == 1992:            
                        value_1993 = data2[col].loc[(data2['YEAR'] == 1993) & (data2['STATE']
                                                                      == state)].values[0]
                        data2[col].loc[(data2['YEAR'] == year) & (data2['STATE']
                                                                      == state)] = value_1993
                        
        except IndexError:
            pass
```

```python
data2['YEAR'].loc[data2.isna().any(axis = 1)].value_counts()
```

We drop the remaining NaN rows.

```python
data2.dropna(inplace=True)
data2.isnull().sum()/data2.isnull().count()
```

### Dealing with outliers

We pick 4 sample states to understand what variables might contain outliers: MI, CO, FL, and NV.

```python
cols = list(data2.columns)
cols.remove('STATE')
cols.remove('PRIMARY_KEY')
```

```python
plt.figure(figsize=(10, 40))

states = ['MICHIGAN', 'COLORADO', 'FLORIDA', 'NEVADA']

for index, col in enumerate(cols):
    plt.subplot(15, 2, index + 1)

    x = [data2[col].loc[data2['STATE'] == state] for state in states]
    
    plt.boxplot(x, labels=states)
    plt.title(col)

plt.tight_layout()
```

We notice that the `AVG_READING_8_SCORE`, `AVG_READING_4_SCORE`, and `STATE_REVENUE` variables might contain outliers, so we explore this further. We plot the values for these variables using Tukey's test.

```python
def tukeys_visualizers(dataframe, list_of_cols):
    
    for col in list_of_cols:
        #print("Calculating outliers for:", col)
        q75, q25 = np.percentile(dataframe[col], [75, 25])
        iqr = q75 - q25

        outliers = []
        thresholds = []

        for threshold in np.arange(1, 15, 0.5):
            min_val = q25 - (iqr*threshold)
            max_val = q75 + (iqr*threshold)

            outlier_count = len((np.where((dataframe[col] > max_val) | (dataframe[col] < min_val))[0]))

            
            outliers.append(outlier_count)
            thresholds.append(threshold)

        visuals = pd.Series(outliers, thresholds)
        visuals.plot(label = col)

    plt.legend()
    plt.xlabel('IQR Threshold')
    plt.ylabel('Number of outliers')
    plt.xticks(ticks = range(1, 15))
```

```python
plt.figure(figsize = (12,95))

for index, state in enumerate(data2['STATE'].unique()):
    plt.subplot(18, 3, index + 1)
    list_of_cols = ['AVG_READING_8_SCORE', 'AVG_READING_4_SCORE', 'STATE_REVENUE']
    
    tukeys_visualizers(data2.loc[data2['STATE'] == state], list_of_cols)
    plt.title(state)
        
plt.show()
plt.tight_layout()
```

It seems as though the notable outliers (IQR threshold > 3) are in `ALASKA`, `COLORADO`, `HAWAII`, `INDIANA`, `MICHIGAN`, `MINNESOTA`, `MISSISSIPPI`, `MICHIGAN`, `NEW MEXICO`, `SOUTH CAROLINA`, `SOUTH DAKOTA` and `TENNESSEE`.
