```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.svm import LinearSVC

%matplotlib inline
```

```python
raw_data = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/epi_r.csv')
raw_data.shape
```

```python
raw_data.head()
```

```python
len(raw_data.columns)
```

```python
raw_data['rating'].hist(bins = 10)
plt.title('Histogram of recipe ratings')
plt.axvline(np.mean(raw_data['rating']), color = 'red', label = 'mean recipe rating')
plt.axvline(np.median(raw_data['rating']), color = 'orange', label = 'median recipe rating')
plt.legend()
plt.show()
```

We see a pretty sharp discontintuity towards the end of higher extreme of the ratings. As such, we'll mark anything above the median value as a 'great recipe', and anything lower as not so great recipe. We are also going to drop rows with missing values in the interest of saving time while training the model.

```python
median_rating = np.median(raw_data['rating'])
data = raw_data.dropna()

data['great_recipe'] = data['rating'].apply(lambda x: x >= median_rating)
data[['rating', 'great_recipe']].head()
```

We'll start by identifying the ingredient columns with the most amount of non-zero data.

```python
sums = data.drop(columns=['title', 'rating', 'calories',
                          'protein', 'fat', 'sodium', 'great_recipe']).sum(axis=0)

sums.sort_values(ascending = False)
```

On the flipside, we'll drop any column that's mentioned less than the 30th percentile of times, which is about 12 times.

```python
drop_thresh = np.percentile(sums.values, q = 30)
drop_thresh
```

```python
drop_cols = sums.loc[sums < drop_thresh].keys()
len(drop_cols)
```

```python
data = data.drop(columns=drop_cols)
data.shape
```

For our baseline model, we'll also identify highly correlated features and temporarily drop them. We start by removing the features that have fall below the 50th percentile (median value) in their absolute correlation with the target variable, `great_recipe`.

```python
corr = data.drop(columns=['title', 'rating', 'calories',
                          'protein', 'fat', 'sodium']).corr()

abs_corr_series = abs(corr['great_recipe']).sort_values(ascending = False)

mean_corr = np.mean(abs_corr_series)
med_corr =  np.median(abs_corr_series)

display(mean_corr, med_corr)
```

```python
drop_cols = abs_corr_series.loc[abs_corr_series < med_corr].keys()
len(drop_cols), drop_cols
```

```python
data = data.drop(columns=drop_cols)
data.shape
```

We now drop the features any are strongly correlated with each other, aka. corr > 0.90.

```python
corr = abs(data.drop(columns=['title', 'rating', 'calories',
                          'protein', 'fat', 'sodium', 'great_recipe']).corr())

upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(np.bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]


data = data.drop(columns=to_drop)
to_drop
```

However, we still have a ton of columns that need to be removed to improve computational efficiency of the features. To do this, we'll start combining similar features as new variables.

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

data_sample = data.sample(frac = 0.4, random_state=13, replace=False)
X = data_sample.drop(columns=['title', 'rating', 'great_recipe'])
Y = data_sample['great_recipe']

X = preprocessing.scale(X)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.33, random_state = 13, stratify = Y)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

## Training the linear model

```python
svm = LinearSVC(verbose=10, random_state=13)
svm.fit(X_train, Y_train)
```

```python
svm.score(X_train, Y_train), svm.score(X_test, Y_test)
```

## Trying other SVM kernels

```python
from sklearn.svm import SVC

for kernel in ['rbf', 'poly', 'sigmoid']:
    svm = SVC(kernel=kernel, verbose = 50, random_state=13)
    svm.fit(X_train, Y_train)
    print(kernel, svm.score(X_train, Y_train), svm.score(X_test, Y_test))
```

Despite varying training scores, the test scores for the four kernels is somewhat consistent. I'll have to tune a given kernel to improve its performance. We can somewhat confidently say that the binary classification isn't a linear function. 

For the purposes of tuning the model, we will assume that the binary classification problem can be represented by either the radial (RBF) or sigmoid kernel, and we will try and tune those two models further.

```python
from sklearn.model_selection import cross_val_score

baseline = SVC(C = 1, kernel = 'rbf', random_state=13, verbose = 100).fit(X_train, Y_train)
baseline_scores = cross_val_score(baseline, X_test, Y_test, cv = 10)
np.mean(baseline_scores), baseline_scores
```

We now tune using GridSearch. Credits to [Geeksforgeeks.org](https://www.geeksforgeeks.org/svm-hyperparameter-tuning-using-gridsearchcv-ml/).




```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf', 'sigmoid'], 
             'random_state': [13]}

grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=100, n_jobs=-1)

# fitting the model for grid search
grid.fit(X_train, Y_train)
```

```python
print(grid.best_estimator_) 
```

```python
from sklearn.metrics import classification_report

print(classification_report(Y_test, grid.predict(X_test)))
```

Looks like none of the models substantially breach the prediction 60% accuracy. I think we might be picking the wrong algorithm here...

```python
from sklearn import ensemble

rfc = ensemble.RandomForestClassifier()
rfc_scores = cross_val_score(rfc, X_train, Y_train, cv = 10)

np.mean(rfc_scores), rfc_scores
```

Well, in this case, we're probably dropping too many variables. Let's go back to the start.

```python
data2 = raw_data.copy()
median_rating = np.median(data2['rating'])
data2 = data2.dropna()

data2['great_recipe'] = data2['rating'].apply(lambda x: x >= median_rating).astype(int)
data2[['great_recipe', 'rating']].head()

sums = data2.drop(columns=['title', 'rating', 'calories',
                          'protein', 'fat', 'sodium', 'great_recipe']).sum(axis=0)

sums.sort_values(ascending = False)
drop_thresh = np.percentile(sums.values, q = 30)
drop_cols = sums.loc[sums < drop_thresh].keys()
data2 = data2.drop(columns=drop_cols)

data_sample2 = data2.sample(frac = 0.4, random_state=13, replace=False)
X = data_sample2.drop(columns=['title', 'rating', 'great_recipe'])
Y = data_sample2['great_recipe']

X = preprocessing.scale(X)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.33, random_state = 13, stratify = Y)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
rfc = ensemble.RandomForestClassifier()
rfc_scores = cross_val_score(rfc, X_train, Y_train, cv = 10)

np.mean(rfc_scores), rfc_scores
```

Strangely adding almost all of the features doesn't help either. Maybe we try adding...all of the features and all of the rows?

```python
data3 = raw_data.copy()
median_rating = np.median(data3['rating'])
data3 = data3.dropna()

data3['great_recipe'] = data3['rating'].apply(lambda x: x >= median_rating).astype(int)

X = data3.drop(columns=['title', 'rating', 'great_recipe'])
Y = data3['great_recipe']

X = preprocessing.scale(X)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.33, random_state = 13, stratify = Y)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
rfc = ensemble.RandomForestClassifier()
rfc_scores = cross_val_score(rfc, X_train, Y_train, cv = 10)

np.mean(rfc_scores), rfc_scores
```

Perhaps unsurprisingly, there's almost no improvement to adding additional features. I think this dataset requires a ton more of feature engineering.
